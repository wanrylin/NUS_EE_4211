## 2.1 Can you explain why you may want to forecast the carpark availability in the future? Who would find this information valuable? What can you do if you have a good forecasting model?

### With the forecasted carpark availability, we can know the information in advance. The advanced information will help us make decision or take proper action. 

### Drivers, government and the carpark management companies will find this information valuable.
Drivers. They can schedule their parking in busy destination before they driving off, and even turn to public transport when facing parking anxiety.

Government. They can use the advanced information for suggestions or reshedule of the new carparks.

Carpark management companies. This information can not only help drivers to avoid choosing parking lots that are already full but also reduce the stress of parking lots with high flow.

### With a good forecasting model, we can provide this model to the government so that they can use it to improve their parking policies. We can also update the precision to the public so that every driver can plan their driving and parking more easily.

Use the month of July 2022 as a training dataset and the month of August 2022 as the test dataset.

(ii) Scatter plot of actual vs predicted hourly values (along with a line showing how good the fit is).

```python
%matplotlib inline

import matplotlib.pyplot as plt
import requests
import json
import pandas as pd
import datetime as dt
import os
import numpy as np
import random
import warnings
import sklearn.linear_model as lm
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures as PF
from sklearn.metrics import r2_score,mean_absolute_error
warnings.filterwarnings("ignore")
```

**Get training dataset**

```python
# grab certain time data and auto-repeat when "time out"
def timegraber_timeout(lstime = None):
    
    # input args processing
    if lstime is None:
        year = 2022
        month = 4
        day = 1
        hour = 0
        minute = 20
        second = 0 
    elif len(lstime) != 6:
        raise IndexError("lenght of list expects 6")
    else:
        year = lstime[0]
        month = lstime[1]
        day = lstime[2]
        hour = lstime[3]
        minute = lstime[4]
        second = lstime[5]
        
        # generate time
    Ddate = dt.datetime(year,month,day,hour,minute,second)

    # step
 #   deltat = dt.timedelta(hour = 1)
    loop = 1

    # update data
    while loop:

        # concatenate datetime
        date_time = f'date_time={str(Ddate.year)}-{str(Ddate.month).zfill(2)}-{str(Ddate.day).zfill(2)}T{str(Ddate.hour).zfill(2)}%3A{str(Ddate.minute).zfill(2)}%3A{str(Ddate.second).zfill(2)}'

        filename = f'EE4211carpark_date_{date_time}.json'
        # do not repeat download and write data
        if os.path.exists(filename) == False:
            # get API site
            site = f'https://api.data.gov.sg/v1/transport/carpark-availability?{date_time}'
            print(site)
            response_API = requests.get(site)
            data = response_API.text
            data = json.loads(data)

            # API may be timed out
            # when time out reget API
            if "message" in data:
                loop = 1
                print("time out")
                continue

            # no timed out then get timestamp
            timestamp = data['items'][0]["timestamp"]
            # get carpark data
            carpark = data['items'][0]["carpark_data"]
            with open(filename, 'w') as fp:
                json.dump(carpark, fp)

        # read file
        df = pd.read_json(filename)
        # reconstuct the frame of data
        for heading in ("total_lots","lot_type","lots_available"):
            df[heading] = df["carpark_info"].apply(lambda x: x[0][heading])
        df = df.drop(["carpark_info"], axis=1)
#        print(filename,"collected")
        loop = 0
        
    return df
```

```python
# The start time of one-month interval
month = 7
day = 1
hour = 0
time_origin = [2022,month,day,hour,20,0]
df_origin = timegraber_timeout(time_origin)
#display(df_origin)

# Change the data to column name
colname = '_'.join(str(i) for i in time_origin)
#print(colname)

# Set up whole dataframe. Rename the column
total_number = df_origin.iloc[:,[0,3]]
df_whole = df_origin.rename(columns={'lots_available':colname})
df_whole = df_whole.drop(df_whole.columns[[1,2]],axis=1)


# Sort the whole dataframe
df_whole = df_whole.sort_values(by=['carpark_number','lot_type'])
#display(df_whole)

# Loop for a month

# gather new data
for day in range(1,32):
    for hour in range(24):
        time_new = [2022,month,day,hour,20,0]
        if time_new == time_origin:
            continue
        df_new = timegraber_timeout(time_new)
        # Change the data to column name
        colname = '_'.join(str(i) for i in time_new)
        #print(colname)
        # Rename the column
        new_number = df_new.iloc[:,[0,2,3]]
        df_new = df_new.rename(columns={'lots_available':colname})
        # Shape the new data
        
        df_new = df_new.drop(df_new.columns[[1,2]],axis=1)
        df_new = df_new.sort_values(by=['carpark_number','lot_type'])
        
        #display(df_new)
        #df_whole = pd.merge(df_whole,df_new,on=['carpark_number','total_lots','lot_type'],how='outer')
        df_whole = pd.merge(df_whole,df_new,on=['carpark_number','lot_type'],how='outer')
        total_number = pd.merge(total_number,new_number,on=['carpark_number','lot_type'],how='outer')
    print('Comparation finished for',day,'day.')
df_whole = df_whole.sort_values(by=['carpark_number','lot_type'])
display(df_whole)
total_number = total_number.sort_values(by=['carpark_number','lot_type'])
display(total_number)
```

```python
# Reset the index of the whole data
print('Origin data:')
df_whole = df_whole.reset_index(drop=True)
display(df_whole)

print('Carparks without data for the most of time: ')
df_wholenan = df_whole[df_whole.isnull().T.any()]
display(df_wholenan)

df_wholereal = df_whole.drop(df_wholenan.index)
df_wholereal = df_wholereal.reset_index(drop=True)
print('Cleaned data: ')
display(df_wholereal)

# Deal with the total lots of each carpark
# Re_index
total_number = total_number.reset_index(drop=True)
# Clean nan
total_number_real = total_number.drop(df_wholenan.index)
total_number_real = total_number_real.reset_index(drop=True)
#display(total_number_real)
# Turn to int and get the one number of total lots of each
total_lots = total_number_real.drop(total_number_real.columns[[0,1]],axis=1).astype(int)
#display(total_lots)
print('The number of total lots of each carpark:')
total_lots_real = total_lots.T.max()
display(total_lots_real)
```

```python
number_carpark = df_wholereal.drop(df_wholereal.columns[[0,1]],axis=1).astype(int)
#display(number_carpark)
number_carpark_sum = number_carpark.sum()
#display(number_carpark_sum)
total_lots_real_sum = total_lots_real.sum()
availrate_sum = number_carpark_sum.div(total_lots_real_sum)

# Plot
fig_sum,ax_sum=plt.subplots(figsize=(10, 5))
ax_sum.plot(availrate_sum)
plt.xlabel('The hours in July 2022 (2022-07-01-0:20 to 2022-07-31-23:20)')
#plt.ylabel('The index of each carpark')
plt.title('The percentage of available lots in whole')
plt.show()
```

**Training model using July data**

```python
def dayfit(deg,weekday,week_sort,y_new):
    if weekday == 1 :
        start = 0
        end = list(week_sort).index(24)
    elif weekday > 1 and weekday < 7 :
        start = list(week_sort).index((weekday - 1) * 24)
        end = list(week_sort).index(weekday * 24)
    elif weekday == 7:
        start = list(week_sort).index((weekday - 1) * 24)
        end = len(week_sort)
        
    certainday = week_sort[start:end]
    certainday_value = y_new[start:end] * 100
    x_tr = np.array(range((weekday - 1) * 24,weekday * 24))
    
    # We create the model.
    lr = lm.LinearRegression()
#     X_poly = PF(degree=deg).fit_transform(certainday)
#     X_tr = PF(degree=deg).fit_transform(np.array(range(start,end)).reshape(-1,1))

    # We train the model on our training dataset.
    lr.fit(np.vander(certainday,deg + 1), certainday_value)
    # Now, we predict points with our trained model.
    y_lr = lr.predict(np.vander(x_tr, deg + 1))
    
    return list(y_lr)
```

```python
y = list(availrate_sum)
week = list(np.linspace(0,7*24-1,7*24))
july = week[4*24:] + week + week + week + week
july = np.array(july).reshape(-1,1)

y_array = np.array(y)
week_sort = np.sort(july[:,0])
y_new = y_array[np.argsort(july[:,0])]
```

```python
predict_rate = []
for i in range(1,8):
    prediction = dayfit(5,i,week_sort,y_new)
    predict_rate += prediction
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 3))
ax.plot(range(0,7*24), predict_rate, '--g')
ax.plot(july, [i * 100 for i in y], 'ok', ms=1)
ax.set_xlim(0, 7*24)
ax.set_ylim(0, 100)
ax.set_title("Avaliable rate for a week")
plt.xlabel("hour")
plt.ylabel("avaliable rate")
plt.grid()
```

```python
# Features,hour，week-day，number of weeks
hour = [0,1,2]
day = [0,1,2,3,4,5,6]
week = [0,1,2,3]

train_x = []
train_y = []
for i in range(744):
    h = int((i%24)/7) + 1
    day = int((i%192)/24) + 1
    week = int((i%768)/192) + 1


    train_x.append([h,day,week])
    train_y.append([new_sum[i]])


train_x = np.array(train_x)
train_y = np.array(train_y)
```
