## Data collecting function

```python
# grab certain time data
def timegraber(lstime = None):
    
    # input args processing
    if lstime is None:
        year = 2022
        month = 4
        day = 12
        hour = 12
        minute = 30
        second = 0 
    elif len(lstime) != 6:
        raise IndexError("lenght of list expects 6")
    else:
        year = lstime[0]
        month = lstime[1]
        day = lstime[2]
        hour = lstime[3]
        minute = lstime[4]
        second = lstime[5]
        
    # generate time
    Ddate = dt.datetime(year,month,day,hour,minute,second)
    # concatenate datetime
    date_time = f'date_time={str(Ddate.year)}-{str(Ddate.month).zfill(2)}-{str(Ddate.day).zfill(2)}T{str(Ddate.hour).zfill(2)}%3A{str(Ddate.minute).zfill(2)}%3A{str(Ddate.second).zfill(2)}'
    
    site = f'https://api.data.gov.sg/v1/transport/carpark-availability?{date_time}'
#     print(site)
    response_API = requests.get(site)
    data = response_API.text
    data = json.loads(data)
    timestamp = data["items"][0]["timestamp"]
    print(timestamp)
    
    filename = f'EE4211carpark_date_{date_time}.json'
    # do not repeat download and write data
    if os.path.exists(filename) == False:
        # file not exist
        data = data["items"][0]["carpark_data"]
        with open(filename, 'w') as fp:
            json.dump(data, fp)
            
    # read file
    df = pd.read_json(filename)
    for heading in ("total_lots","lot_type","lots_available"):
        df[heading] = df["carpark_info"].apply(lambda x: x[0][heading])
    df = df.drop(["carpark_info"], axis=1)
    
    return df
```

### Find the update frequency of the carpark data

```python
# Change these values for different dates
year = 2022
month = 4
day = 12
hour = 12
minute = 30
second = 0

# generate time
Ddate = dt.datetime(year,month,day,hour,minute,second)

# step
deltat = dt.timedelta(minutes = 1)
loop = 0

# update data
while True:
    loop += 1
    
    # concatenate datetime
    date_time = f'date_time={str(Ddate.year)}-{str(Ddate.month).zfill(2)}-{str(Ddate.day).zfill(2)}T{str(Ddate.hour).zfill(2)}%3A{str(Ddate.minute).zfill(2)}%3A{str(Ddate.second).zfill(2)}'
    
    filename = f'EE4211carpark_date_{date_time}.json'
    # do not repeat download and write data
    if os.path.exists(filename) == False:
        # get API site
        site = f'https://api.data.gov.sg/v1/transport/carpark-availability?{date_time}'
        print(site)
        response_API = requests.get(site)
        data = response_API.text
        data = json.loads(data)
        
        # API may be timed out
        # when time out reget API
        if "message" in data:
            loop -= 1
            print("time out")
            continue
        
        # no timed out then get timestamp
        timestamp = data['items'][0]["timestamp"]
        # get carpark data
        carpark = data['items'][0]["carpark_data"]
        with open(filename, 'w') as fp:
            json.dump(carpark, fp)
   
    # read file
    df = pd.read_json(filename)
    # reconstuct the frame of data
    for heading in ("total_lots","lot_type","lots_available"):
        df[heading] = df["carpark_info"].apply(lambda x: x[0][heading])
    df = df.drop(["carpark_info"], axis=1)
    
    if loop == 1:
        # store the old df
        df_old = df.copy(deep = True)
        # add delta time
        Ddate = Ddate + deltat
        continue
    
    #
    # compare update_datetime and break when data updated
    #
    # compare the length of update time first
    update = False
    if len(df.update_datetime) == len(df_old.update_datetime):
        df3 = (df.update_datetime == df_old.update_datetime).any(axis = None)
    else:
        update = True
    
    # compare the value next
    if df3 == True and update == False:
        # add delta time
        Ddate = Ddate + deltat
        # store old df
        df_old = df.copy(deep = True)
        print(Ddate, df3)
    else:
        length = np.min([len(df.update_datetime),len(df_old.update_datetime)])
        # print the location of the different update time
        upt_diff = (df.update_datetime[0:length] != df_old.update_datetime[0:length])
        loc_up = upt_diff[upt_diff.values == True].index
        print('location of the upload',loc_up)
        print(df_old.update_datetime[loc_up],df.update_datetime[loc_up])
        break
```

### A function that repeat collecting data when "time out":

```python
# grab certain time data and auto-repeat when "time out"
def timegraber_timeout(lstime = None):
    
    # input args processing
    if lstime is None:
        year = 2022
        month = 4
        day = 1
        hour = 0
        minute = 20
        second = 0 
    elif len(lstime) != 6:
        raise IndexError("lenght of list expects 6")
    else:
        year = lstime[0]
        month = lstime[1]
        day = lstime[2]
        hour = lstime[3]
        minute = lstime[4]
        second = lstime[5]
        
        # generate time
    Ddate = dt.datetime(year,month,day,hour,minute,second)

    # step
 #   deltat = dt.timedelta(hour = 1)
    loop = 1

    # update data
    while loop:

        # concatenate datetime
        date_time = f'date_time={str(Ddate.year)}-{str(Ddate.month).zfill(2)}-{str(Ddate.day).zfill(2)}T{str(Ddate.hour).zfill(2)}%3A{str(Ddate.minute).zfill(2)}%3A{str(Ddate.second).zfill(2)}'

        filename = f'EE4211carpark_date_{date_time}.json'
        # do not repeat download and write data
        if os.path.exists(filename) == False:
            # get API site
            site = f'https://api.data.gov.sg/v1/transport/carpark-availability?{date_time}'
            print(site)
            response_API = requests.get(site)
            data = response_API.text
            data = json.loads(data)

            # API may be timed out
            # when time out reget API
            if "message" in data:
                loop = 1
                print("time out")
                continue

            # no timed out then get timestamp
            timestamp = data['items'][0]["timestamp"]
            # get carpark data
            carpark = data['items'][0]["carpark_data"]
            with open(filename, 'w') as fp:
                json.dump(carpark, fp)

        # read file
        df = pd.read_json(filename)
        # reconstuct the frame of data
        for heading in ("total_lots","lot_type","lots_available"):
            df[heading] = df["carpark_info"].apply(lambda x: x[0][heading])
        df = df.drop(["carpark_info"], axis=1)
#        print(filename,"collected")
        loop = 0
        
    return df
```

### Loop for the one month data collection

```python
import warnings
warnings.filterwarnings("ignore")

# The start time of one-month interval
month = 4
day = 1
hour = 0
time_origin = [2022,month,day,hour,20,0]
df_origin = timegraber_timeout(time_origin)
#display(df_origin)

# Change the data to column name
colname = '_'.join(str(i) for i in time_origin)
#print(colname)

# Set up whole dataframe. Rename the column
total_number = df_origin.iloc[:,[0,3]]
df_whole = df_origin.rename(columns={'lots_available':colname})
df_whole = df_whole.drop(df_whole.columns[[1,2]],axis=1)


# Sort the whole dataframe
df_whole = df_whole.sort_values(by=['carpark_number','lot_type'])
#display(df_whole)

# Loop for a month

# gather new data
for day in range(1,31):
    for hour in range(24):
        time_new = [2022,month,day,hour,20,0]
        if time_new == time_origin:
            continue
        df_new = timegraber_timeout(time_new)
        # Change the data to column name
        colname = '_'.join(str(i) for i in time_new)
        #print(colname)
        # Rename the column
        new_number = df_new.iloc[:,[0,2,3]]
        df_new = df_new.rename(columns={'lots_available':colname})
        # Shape the new data
        
        df_new = df_new.drop(df_new.columns[[1,2]],axis=1)
        df_new = df_new.sort_values(by=['carpark_number','lot_type'])
        
        #display(df_new)
        #df_whole = pd.merge(df_whole,df_new,on=['carpark_number','total_lots','lot_type'],how='outer')
        df_whole = pd.merge(df_whole,df_new,on=['carpark_number','lot_type'],how='outer')
        total_number = pd.merge(total_number,new_number,on=['carpark_number','lot_type'],how='outer')
    print('Comparation finished for',day,'day.')
df_whole = df_whole.sort_values(by=['carpark_number','lot_type'])
display(df_whole)
total_number = total_number.sort_values(by=['carpark_number','lot_type'])
display(total_number)
```

# Proposal

Currently, parking problem has been an important concern of families with cars. One of the most vital reasons that a lot of people refuse to go out by driving is that it’s hard to find a proper parking lot. From the visualization of the availability of each carpark, we find that there are some regularities based on working hours, holidays and geographical areas. During off-duty hours and public holidays, the HDB carparks seem to have fewer available lots. So, to better solve the parking problem, we proposed three research methods using HDB parking lots dataset.

First, suggestions for the location of new HDB carparks. In the aspect of urban planning, the parking problem is severe in some area while in other areas, the parking problem is not that severe. We could use the dataset to analyze the saturation of parking lots in an area. And the saturation index of the parking lots in an area can be used to further analyze if this area needs more parking lots to reduce the stress of parking problem. For instance, we can provide the government the areas with the highest saturation index so that government can decide if they need to build more parking lots in these areas. To do this, we can first classify the parking lots into different area, then, calculate the saturation index in every area, and finally, provide the areas with the highest saturation index.

Second, carparks availability prediction. We can use the dataset to predict the availability of a parking lot in a particular period. This can not only help drivers to avoid choosing parking lots that are already full but also reduce the stress of parking lots with high flow. There are two ways to predict the availability of a parking lot, firstly, we may only use the data of this particular parking lot; secondly, we may use the data of the parking lots in the near area because the availability of parking lots in a specific area shows some correlation which may help to improve the performance of our prediction model.

Third, parking scheduling. A lot of drivers are facing parking anxiety because of the lack of parking place in their destination. We can use the dataset to give drivers the real-time information of nearby parking lots and suggest them with the best alternative parking lot. Through the parking lot location file from Data.gov.sg, we can know the precise location of every parking lot, thus, if drivers provide their destinations, we can suggest them the nearest available alternative parking lots. And combined with the second method, with the destinations and the expected arrival time, we can also give drivers suggestions of nearby available alternative parking lots in the future
