## Data collecting blocks. No need to run because they were download in advanced with csv.

```python
# grab certain time data and auto-repeat when "time out"
def timegraber_timeout(lstime = None):
    
    # input args processing
    if lstime is None:
        year = 2022
        month = 4
        day = 1
        hour = 0
        minute = 20
        second = 0 
    elif len(lstime) != 6:
        raise IndexError("lenght of list expects 6")
    else:
        year = lstime[0]
        month = lstime[1]
        day = lstime[2]
        hour = lstime[3]
        minute = lstime[4]
        second = lstime[5]
        
        # generate time
    Ddate = dt.datetime(year,month,day,hour,minute,second)

    # step
 #   deltat = dt.timedelta(hour = 1)
    loop = 1

    # update data
    while loop:

        # concatenate datetime
        date_time = f'date_time={str(Ddate.year)}-{str(Ddate.month).zfill(2)}-{str(Ddate.day).zfill(2)}T{str(Ddate.hour).zfill(2)}%3A{str(Ddate.minute).zfill(2)}%3A{str(Ddate.second).zfill(2)}'

        filename = f'EE4211carpark_date_{date_time}.json'
        # do not repeat download and write data
        if os.path.exists(filename) == False:
            # get API site
            site = f'https://api.data.gov.sg/v1/transport/carpark-availability?{date_time}'
            print(site)
            response_API = requests.get(site)
            data = response_API.text
            data = json.loads(data)

            # API may be timed out
            # when time out reget API
            if "message" in data:
                loop = 1
                print("time out")
                continue

            # no timed out then get timestamp
            timestamp = data['items'][0]["timestamp"]
            # get carpark data
            carpark = data['items'][0]["carpark_data"]
            with open(filename, 'w') as fp:
                json.dump(carpark, fp)

        # read file
        df = pd.read_json(filename)
        # reconstuct the frame of data
        for heading in ("total_lots","lot_type","lots_available"):
            df[heading] = df["carpark_info"].apply(lambda x: x[0][heading])
        df = df.drop(["carpark_info"], axis=1)
#        print(filename,"collected")
        loop = 0
        
    return df
```

### Getting July data

```python
# The start time of one-month interval
month = 7
day = 1
hour = 0
time_origin = [2022,month,day,hour,20,0]
df_origin = timegraber_timeout(time_origin)
#display(df_origin)

# Change the data to column name
colname = '_'.join(str(i) for i in time_origin)
#print(colname)

# Set up whole dataframe. Rename the column
total_number = df_origin.iloc[:,[0,3]]
df_whole = df_origin.rename(columns={'lots_available':colname})
df_whole = df_whole.drop(df_whole.columns[[1,2]],axis=1)


# Sort the whole dataframe
df_whole = df_whole.sort_values(by=['carpark_number','lot_type'])
#display(df_whole)

# Loop for a month

# gather new data
for day in range(1,32):
    for hour in range(24):
        time_new = [2022,month,day,hour,20,0]
        if time_new == time_origin:
            continue
        df_new = timegraber_timeout(time_new)
        # Change the data to column name
        colname = '_'.join(str(i) for i in time_new)
        #print(colname)
        # Rename the column
        new_number = df_new.iloc[:,[0,2,3]]
        df_new = df_new.rename(columns={'lots_available':colname})
        # Shape the new data
        
        df_new = df_new.drop(df_new.columns[[1,2]],axis=1)
        df_new = df_new.sort_values(by=['carpark_number','lot_type'])
        
        #display(df_new)
        #df_whole = pd.merge(df_whole,df_new,on=['carpark_number','total_lots','lot_type'],how='outer')
        df_whole = pd.merge(df_whole,df_new,on=['carpark_number','lot_type'],how='outer')
        total_number = pd.merge(total_number,new_number,on=['carpark_number','lot_type'],how='outer')
    print('Comparation finished for',day,'day.')
july_whole = df_whole.sort_values(by=['carpark_number','lot_type'])
display(july_whole)
july_total_lots = total_number.sort_values(by=['carpark_number','lot_type'])
display(july_total_lots)
```

### Clean the data: july_wholereal, july_lots, aug_wholereal, aug_lots

```python
#JULY

print('Carparks without data for the most of time in July: ')
july_wholenan = july_whole[july_whole.isnull().T.any()]
#display(july_wholenan)

july_wholereal = july_whole.drop(july_wholenan.index)
july_wholereal = july_wholereal.reset_index(drop=True)
print('Cleaned data: ')
display(july_wholereal)
#july_wholereal.to_csv('july_wholereal.csv')

# Deal with the total lots of each carpark
# Clean nan
july_lots_real = july_total_lots.drop(july_wholenan.index)
july_lots_real = july_lots_real.reset_index(drop=True)
#display(july_lots_real)

# Turn to int and get the one number of total lots of each
july_lots = july_lots_real.drop(july_lots_real.columns[[0,1]],axis=1).astype(int)
#display(july_lots)
print('The number of total lots of each carpark:')
july_lots = july_lots.T.max()
july_lots[july_lots == 0] = 1
display(july_lots)

#AUG

print('Carparks without data for the most of time in Aug: ')
aug_wholenan = aug_whole[aug_whole.isnull().T.any()]
#display(aug_wholenan)

aug_wholereal = aug_whole.drop(aug_wholenan.index)
aug_wholereal = aug_wholereal.reset_index(drop=True)
print('Cleaned data: ')
display(aug_wholereal)
#aug_wholereal.to_csv('aug_wholereal.csv')

# Deal with the total lots of each carpark
# Clean nan
aug_lots_real = aug_total_lots.drop(aug_wholenan.index)
aug_lots_real = aug_lots_real.reset_index(drop=True)
#display(aug_lots_real)

# Turn to int and get the one number of total lots of each
aug_lots = aug_lots_real.drop(aug_lots_real.columns[[0,1]],axis=1).astype(int)
#display(aug_lots)
print('The number of total lots of each carpark:')
aug_lots = aug_lots.T.max()
aug_lots[aug_lots == 0] = 1
display(aug_lots)
```

```python
# JULY Availablity

july_div = july_wholereal.drop(july_wholereal.columns[[0,1]],axis=1).astype(int)
july_div = july_div.div(july_lots,axis=0)
july_div[july_div>1] = 1
display(july_div)

# AUG Availablity

aug_div = aug_wholereal.drop(aug_wholereal.columns[[0,1]],axis=1).astype(int)
aug_div = aug_div.div(aug_lots,axis=0)
aug_div[aug_div>1] = 1
display(aug_div)
```
